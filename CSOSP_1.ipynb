```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform
from scipy.stats import expon
from scipy.stats import variation
from scipy.stats import norm
```

```python
### Generate exponential distributed random variables given the mean 
### and number of random variables
def exponential_inverse_trans(n = 1,mean = 1):
    U = uniform.rvs(size = n)
    X = - mean * np.log(1 - U)
    actual = expon.rvs(size = n, scale = mean)
    
    plt.figure(figsize = (8,5))
    plt.hist(X, bins = 50, alpha = 0.5, label = "Generated r.v.")
    plt.hist(actual, bins = 50, alpha = 0.5, label = "Actual r.v.")
    plt.title("Generated vs Actual %i Exponential Random Variables" %n)
    plt.legend()
    plt.show()
    return X
```

```python
def pareto_inverse_trans(n = 1, lamb = 1, alpha = 1):
    U = uniform.rvs(size = n)
    X = lamb * (1 / (1 - U)**(1 / alpha) - 1)
    
    plt.figure(figsize = (4,2))
    plt.hist(X, bins = 50, alpha = 0.5, label = "Generated r.v.")
    plt.title("Generated %i Pareto Random Variables, lambda = %.1f, alpha = %.1f" %(n, lamb, alpha))
    plt.legend()
    plt.show()
    return X
```

```python
N = [12, 50, 100, 500, 1000, 1500]
n = 1000

mu = np.divide(1, 2)
var = np.divide(1, 12)
    
for j in N:
    X = []
    means = []

    for i in range(j):
        U = uniform.rvs(size = n)
        X.append(U)
        means.append(U.mean())

    Z = np.sqrt(n) * (means - mu) / np.sqrt(var)
    norm_dens = norm.rvs(size = n, loc = 0, scale = 1)

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(7, 2))
    fig.suptitle('For N = %i' %j)
    
    ax1.hist(Z, bins = 11, density = True, alpha = 0.5, label = "Z")
    ax1.hist(norm_dens, bins = 11, density = True, alpha = 0.5, label = "Norm")
    ax1.legend()
    
    ax2.hist(Z, bins = 11, cumulative = True, density = True, alpha = 0.5, label = "Z")
    ax2.hist(norm_dens, bins = 11, cumulative = True, density = True, alpha = 0.5, label = "Norm")
    ax2.legend()
```
