```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform
```

# Simulation of stable random variable

$\Theta$ and $W$ - independent;

$\Theta$ uniformly distributed on $\left(-\frac{\pi}{2}, \frac{\pi}{2} \right)$; 

$W$ exponentially distributed with mean 1;

```python
n = 1000
mean = 1

Theta = uniform.rvs(size = n, loc = - np.pi / 2, scale = np.pi)
U = uniform.rvs(size = n)
W = - mean * np.log(1 - U)
```

```python
def plot(X):
    plt.figure(figsize = (4,3))
    plt.hist(X, bins = 15, label = "Generated r.v.")
    plt.title("Generated Stable Random Variables with alpha = %.2f" %alpha)
    plt.legend()
    plt.show()
```

```python
def cdf(X):
    count, bins_count = np.histogram(X, bins = 15)
    pdf = count / sum(count)
    cdf = np.cumsum(pdf)
    
    plt.figure(figsize = (4,3))
    plt.title("CDF and PDF with alpha = %.2f" %alpha)
    plt.plot(bins_count[1:], pdf, label = 'PDF', alpha = 0.5)
    plt.plot(bins_count[1:], cdf, label = 'CDF', alpha = 0.5)
    plt.legend()
    plt.show()
    
```

# Symmetric case

When $\alpha = 1$

```python
alpha = 1
Z_sym = np.tan(Theta)
```

```python
plot(Z_sym)
cdf(Z_sym)
```

When $\alpha \neq 1$

```python
alpha = 1.7
Z_sym_alp = np.sin(alpha*Theta) / (np.cos(Theta))**(1/alpha) * ((np.cos((alpha - 1) * Theta)) / W)**((1-alpha)/alpha)
```

```python
plot(Z_sym_alp)
cdf(Z_sym_alp)
```

# Nonsymmetric case

When $\alpha = 1$

```python
alpha = 1
beta = 0
Z_nonsym = 2 / np.pi * ((np.pi / 2 + beta * Theta) * np.tan(Theta) - beta * np.log(np.pi / 2 * W * np.cos(Theta) / (np.pi / 2 + beta * Theta)))
```

```python
plot(Z_nonsym)
cdf(Z_nonsym)
```

When $\alpha \neq 1$

```python
alpha = 0.5
beta = 0.5
Theta_0 = np.arctan(beta * np.tan(np.pi * alpha / 2)) / alpha

Z_nonsym_alp = np.sin(alpha * (Theta_0 + Theta)) / (np.cos(alpha * Theta_0) * np.cos(Theta))**(1/alpha) * ((np.cos(alpha * Theta_0 + (alpha - 1) * Theta)) / W)**((1-alpha)/alpha)
```

```python
plot(Z_nonsym_alp)
cdf(Z_nonsym_alp)
```
