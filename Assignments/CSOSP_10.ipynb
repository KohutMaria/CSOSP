# Lamparti transformation

# Levy processes

```python
def simulate_process(Z, n, delta):
    X = [0]
    for k in range(n-1):
        X.append(X[k] + Z[k])
    return X
```

## Brownian motion

```python
def sim_br(t):
    # simulate Brownian motion
    incr = np.diff(t)
    Z = [np.random.normal(0, np.sqrt(incr[0]), 1)]
    for dt in incr:
        Z.append(np.random.normal(0, np.sqrt(dt), 1))
    return np.cumsum(Z)
```

```python
n = 1000 
delta = 0.5
```

```python
t_br = [k * delta for k in range(n)]
X_br = sim_br(t_br)

plt.plot(t_br, X_br)
plt.title('Brownian motion')
plt.show()
```

## Lamparti transformation using Brownian motion

$X(t), \; t >= 0$ - Brownian motion

$Y(y) = e^{-t/2} X(e^t), \; t \in R$ - stationary Ornstein-Uhlenbeck process 

```python
def lamp_transf(t):
    e_t = np.exp(t)
    X = sim_br(e_t)
    Y = np.exp((-t / 2)) * X
    return Y
```

```python
t = np.arange(-100, 100)
```

```python
Y = lamp_transf(t)
plt.plot(t, Y)
plt.title('Stationary Ornstein-Uhlenbeck process')
plt.show()
```
