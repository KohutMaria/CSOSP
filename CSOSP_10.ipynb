# Levy processes

```python
def simulate_process(Z, n, delta):
    X = [0]
    for k in range(n-1):
        X.append(X[k] + Z[k])
    return X
```

## Brownian motion

```python
def sim_br(t):
    incr = np.diff(t)
    Z = [np.random.normal(0, np.sqrt(incr[0]), 1)]
    for dt in incr:
        Z.append(np.random.normal(0, np.sqrt(dt), 1))
    return np.cumsum(Z)
```

```python
def lamp_transf(t):
    e_t = np.exp(t)
    X = sim_br(e_t)
    Y = np.exp((-t / 2)) * X
    return Y
```
